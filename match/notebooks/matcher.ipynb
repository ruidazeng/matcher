This notebook is meant for comparing two impedance signatures given by absolute filepaths. The command outputs the percent match in the bitwise representations of the two specified files. 

```python
# view files
print(os.listdir(os.path.join(os.getcwd(), "VT")))
print(os.listdir(os.path.join(os.getcwd(), "TT")))
print(os.listdir(os.path.join(os.getcwd(), "test")))
print(os.listdir(os.path.join(os.getcwd(), "multi")))
print(os.listdir(os.path.join(os.getcwd(), "orig")))
```

```python
# absolute filepaths expected
ospj = os.path.join
cwd = os.getcwd()
file1 = ospj(cwd, "TT", "QR1_2.txt")
file2 = ospj(cwd, "VT", "QR1_2.csv")

SIZE = 200 # should be in [1, 4000]
```

```python
# read data and encode
freq1, imped1, width1 = readData(file1)
freq2, imped2, width2 = readData(file2)

impAlign1, impAlign2 = alignSignature(imped1, imped2, freq1, freq2, SIZE)
width = bitWidth(width1, width2)

# convert aligned signatures to binary (todo add other representations)
sig1, integerImpedance1 = signatureToBinary(impAlign1, width)
sig2, integerImpedance2 = signatureToBinary(impAlign2, width)
```

```python
# compute distance
hammingDistance, percentMatchingBits = getHammingDistance(sig1, sig2)
```

```python
print(f"Comparing {file1} and {file2}")
print(f"Encoding length: {width} bits")
print(f"Hamming distance: {hammingDistance}/{len(sig1)} bits = {percentMatchingBits}% match")
```
